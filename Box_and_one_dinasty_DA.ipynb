### Descrição dos headers

Esta tabela está com os números Totais dos jogadores em toda a temporada de 23-24 da NBA. A descrição das siglas das colunas a seguir:


PLAYER_ID      ID do jogador

RANK           Classificação segundo a NBA

PLAYER         Nome do Jogador

TEAM_ID        ID do Time

TEAM           Time em que jogou na temporada

GP             Quantidade de jogos jogados na temporada

MIN            Quantidade total de minutos jogados

FGM            Arremessos convertidos

FGA            Arremessos tentados

FG_PCT         Relação entre arremessos convertidos e tentados

FG3M           Arremessos de 3 convertidos

FG3A           Arremessos de 3 tentados

FG3_PCT        Relação entre os arremessos de 3 convertidos e tentados

FTM            Lances livres convertidos

FTA            Lances livres tentados

FT_PCT         Relação entre os lances livres convertidos e tentados

OREB           Rebotes ofensivos pegos

DREB           Rebotes defensivos pegos

REB            Rebotes pegos

AST            Assistências

STL            Roubos de bola

BLK            Bloqueios

TOV            Perda de bola

PF             Faltas cometidas

PTS            Pontos

EFF            Eficiência (PTS + REB + AST + STL + BLK − Missed FG − Missed FT - TO) / GP

AST_TOV        Relação entre assistências/perdas de bola

STL_TOV        Relação entre roubos de bola/perdas de bola

OWNER          Time da liga de Fantasy que tem o jogador


A NBA não fornece as estatísticas de todos os jogadores por jogo pois alguns deles não jogam uma quantidade de minutos suficientes ou jogam muitos poucos jogos por temporada, tornando muito voláteis seus valores e dificultando estatísticas

### Transformação de Totais para por Jogo

```python
# Novo dataset com valores per game (pg)
box_pg = box.copy()
```

```python
# Colunas a ser transformadas
colunas_porjogo = [
    'MIN', 'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
    'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
# Estrutura de repetição para fazer o calculo da média com base nos GP
for col in colunas_porjogo:
    box_pg[col] = (box_pg[col] / box_pg['GP']).round(1)
```

```python
box_pg.dtypes
```

```python
#Remover coluna eff
box_pg = box_pg.drop(columns=['EFF'])
```

```python
#Conferindo conversão
box_pg.sort_values(by='RANK')
```

### Sleepers

Sleeper é um termo em ligas de fantasia para designar um jogador que tem potencial a qualquer momento a estourar na liga e aumentar bastante a sua margem de estatíscias. Este fenomeno ocorre por diversos motivos, como o jogador que compete a posição com ele foi trocado ou se lesionou, ele amadurece depois de alguns anos de adaptação na liga (terceiro ano de novatos, auge da carreira que geralmente é aos 27-29 anos).

Logo, procuraremos jogadores com maior diferença de rank entre o rank atual da nba e o seu Z score e a posição em que ele foi draftado e o OWNER que escolheu este jogador.

### Adicionando Z-Scores do Basketball Monster (BBM)

```python
#Importação de Z-Scores
zscores = pd.read_csv('/content/drive/MyDrive/NBA/Zscores/zscores.csv')
```

```python
#Transformar zscores em dataframe type
zscores = pd.DataFrame(zscores)
```

```python
#Visualização do dataframe
zscores.head(1)
```

```python
#Renomear "value" para "zscore"
zscores.rename(columns={'Value': 'Zscore'}, inplace=True)
zscores.rename(columns={'Rank': 'BBM_RANK'}, inplace=True)
```

```python
# Selecionar apenas a coluna 'zscore' do dataframe zscores para dar merge com box_pg
zscores_subset = zscores[['Name', 'Zscore', 'BBM_RANK']]
```

```python
#Verificando tipos em zscores_subset
zscores_subset.dtypes
```

```python
#Transformando Zscore para float convertendo virgulas para pontos
zscores_subset['Zscore'] = zscores_subset['Zscore'].str.replace(',', '.').astype(float)
```

```python
#zscore com apenas 2 casas decimais
zscores_subset['Zscore'] = zscores_subset['Zscore'].round(2)
```

```python
#Assegurar que tanto zscores_subset quanto box_pg são DataFrames do pandas
zscores_subset = pd.DataFrame(zscores_subset)
box_pg = pd.DataFrame(box_pg)
```

```python
#Deixando chave principal e chave estrangeira com mesmo nome
zscores_subset.rename(columns={'Name': 'PLAYER'}, inplace=True)
```

```python
# Realizar o merge usando a coluna "PLAYER" como chave
box_df = pd.merge(box_pg, zscores_subset, on='PLAYER')
```
